# Data Preprocessing

## Handle missing values


```python

updated_df = updated_df.dropna(subset=['likes', 'num_comments', 'followers'])  # Drop rows with missing likes, comments, or followers
updated_df = updated_df.drop_duplicates(subset=['post_id'])  # Remove duplicates
updated_df = updated_df.drop(['has_handshake','discovery_input','photos',	'videos','profile_image_link','shortcode','pk','thumbnail','partnership_details','profile_url','description'],axis=1)

```

## Check  for null values


```python
updated_df.isnull().sum()
```

## Replacing null values

```python

for i in updated_df.select_dtypes("number").columns:
  updated_df[i] = updated_df[i].fillna(updated_df[i].mean())
for i in updated_df.select_dtypes("object").columns:
  updated_df[i] = updated_df[i].fillna(updated_df[i].mode()[0])


updated_df.isnull().sum()
```

```python
updated_df.head()
```

## Processing "coauther_producer" columns to get their names

```python
updated_df['coauthor_producers'] = updated_df['coauthor_producers'].str.replace('[', '').str.replace(']', '').str.replace("'", '').str.strip() # to get the values inside the list which is in string format
updated_df['coauthor_producers']  =  updated_df['coauthor_producers'].replace(r'^\s*$', np.nan, regex=True)
updated_df['coauthor_producers'].isnull().sum() # here maximum values are null after replacing empty lists with null , hence we will delete this column
updated_df.drop(['coauthor_producers','url','display_url',],axis=1,inplace=True) # the columns with urls are deleted as they are expired and contain no info inside it
```

## Processing the column latest comments to create new columns from its content

```python
updated_df['latest_comments'] =  updated_df['latest_comments'].apply(lambda x :str.strip(x,'[]')).to_dict()
```

```python
updated_df['latest_comments'][0:100]
```

```python
updated_df['latest_comments'] = updated_df['latest_comments'].apply(lambda x :str.strip(x,"{}"))
```

```python
import ast

# Parse each row in the 'latest_comments' column as a dictionary
def parse_latest_comments(comment_str):
    # Add curly braces to make it a valid dictionary format
    comment_str = '{' + comment_str + '}'
    # Convert the string to a dictionary
    try:
        return ast.literal_eval(comment_str)
    except (ValueError, SyntaxError):
        return {"comments": None, "likes": None, "user_commenting": None}

# Apply the function to parse the strings
df_parsed = updated_df['latest_comments'].apply(parse_latest_comments)

# Convert parsed dictionaries into a DataFrame and align columns
parsed_df = pd.json_normalize(df_parsed)  # Creates a new DataFrame from dictionary keys

# Concatenate parsed columns with the original DataFrame
updated_df = pd.concat([updated_df, parsed_df], axis=1)

# Display the DataFrame with new columns
updated_df
```

```python
updated_df.isnull().sum() # after processing we understand that most of the values of the newly created columns are empty hence we will delete these columns along with latest-comments column
```

```python
updated_df.drop(['comments',"user_commenting","replies","latest_comments"],axis=1,inplace=True)
updated_df.drop(['likes'],axis=1,inplace=True)
```

## Time Series Analysis

decomposition plot

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.subplots as sp
import plotly.graph_objs as go

# Decompose the time series data
decomposition = seasonal_decompose(date_engage.set_index('date')['engagement_score_view'], model='additive', period=30)
trend, seasonal, residual = decomposition.trend, decomposition.seasonal, decomposition.resid

# Create subplots for each component (trend, seasonal, residual)
fig = sp.make_subplots(
    rows=3, cols=1,
    subplot_titles=["Trend Component", "Seasonal Component", "Residual Component"],
    shared_xaxes=True,
    vertical_spacing=0.1
)

# Add traces for trend, seasonal, and residual components
fig.add_trace(go.Scatter(x=date_engage['date'], y=trend, mode='lines', name='Trend'),
              row=1, col=1)

fig.add_trace(go.Scatter(x=date_engage['date'], y=seasonal, mode='lines', name='Seasonal'),
              row=2, col=1)

fig.add_trace(go.Scatter(x=date_engage['date'], y=residual, mode='lines', name='Residual'),
              row=3, col=1)

# Update the layout with labels and title
fig.update_layout(
    title="Decomposition of Engagement Score Over Time",
    xaxis_title="Date",
    yaxis_title="Engagement Score",
    height=800
)

fig.show()
```

## Creating new column so as to analyze and find patterns in content posting to get higher engagement rate and views

```python
updated_df['month'] = updated_df['date'].dt.month
updated_df['day'] = updated_df['date'].dt.day
updated_df['hour'] = updated_df['date'].dt.hour
updated_df['minute'] = updated_df['date'].dt.minute
```

```python
updated_df.head()
```

```python
# Creating bins of hours and minutes separately
hour_bins = [0, 3, 7, 11, 15, 19, 23]
hour_labels = ['0-3', '4-7', '8-11', '12-15', '16-19', '20-23']
minute_bins = [0, 14, 29, 44, 59]
minute_labels = ['0-14', '15-29', '30-44', '45-59']

# Create bins for hour and minute columns
updated_df['hour_bin'] = pd.cut(updated_df['hour'], bins=hour_bins, labels=hour_labels, right=True)
updated_df['minute_bin'] = pd.cut(updated_df['minute'], bins=minute_bins, labels=minute_labels, right=True)

# Group by the bins and calculate the mean engagement score, views, and video play count
result = updated_df.groupby(['hour_bin']).agg({
    'engagement_score_view': 'mean',

}).reset_index().sort_values(by='engagement_score_view',ascending=False)

# Display the result
px.histogram(result,x='hour_bin',y='engagement_score_view',title='Engagement Score Over Hours')
```

So , according the hours of content posting the engagement_score_view is the highest between 8-11 am and then followed by 4-7 pm. Content posters should try posting in these times to gain higher engagement

```python
# Sorting min_bin according to engaement score
minute_bin_engagement = result = updated_df.groupby(['minute_bin']).agg({
    'engagement_score_view': 'mean',

}).reset_index().sort_values(by='engagement_score_view',ascending=False)

px.bar(minute_bin_engagement,x='minute_bin',y='engagement_score_view',title='Engagement Score Over Minutes')
```

30 -44 mins of the hour clock get the highest engagement score

```python
## Checking engagement score according to day posted
date_engage = updated_df.groupby('day').agg({'engagement_score_view': 'sum'}).reset_index()
fig_date_engagement = px.bar(date_engage, x='day', y='engagement_score_view',
                              title='Engagement Score Over Day')
fig_date_engagement.show()
```
